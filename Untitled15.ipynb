```python
import xml.etree.ElementTree as ET
import glob
import matplotlib.pyplot as plt
```

```python
files = glob.glob('/content/train/train/Final Train Dataset/*.xml') + glob.glob('/content/train/train/Final Train Dataset/[0-9]*.xml')
vehicle_type = {}
# testing code below 
for i in range(len(files)):
  try:
    tree = ET.parse(files[i])
  except:
    pass
  root = tree.getroot()

  for child in root.findall('object'):
    if child.find('name').text in vehicle_type:
      vehicle_type[child.find('name').text] += 1
    else:
      vehicle_type[child.find('name').text] = 1

plt.bar(range(len(vehicle_type)), list(vehicle_type.values()), align='center')
plt.xticks(range(len(vehicle_type)), list(vehicle_type.keys()),rotation='vertical')
```
